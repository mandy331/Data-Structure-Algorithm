## Stack Introduction

+ Stack是具有「Last-In-First-Out」的資料結構(可以想像成用紙箱裝書)，「最晚進入Stack」的書會「最先被取出」，「最早進入Stack」的資料則「最晚被取出」。Stack是一個抽象的概念，只要符合「Last-In-First-Out」的資料結構，都可以視為Stack。


+ Stack的應用是：
    + 編輯器回到Undo的功能。
    + 網頁瀏覽的「回到上一頁」。
    + DFS以Stack改寫。

## Stack的功能：

- Push(data)：把資料放進Stack。
    + 把書放進箱子。
- Pop：把「最上面」的資料從Stack中移除。
    + 把位於箱子「最上面」的書拿出來。
- Top：回傳「最上面」的資料，不影響資料結構本身。
    + 確認目前位於箱子「最上面」的是哪本書。
- IsEmpty：確認Stack裡是否有資料，不影響資料結構本身。
    + 確認箱子裡面有沒有書。
- getSize：回傳Stack裡的資料個數，不影響資料結構本身。
    + 記錄目前箱子已經裝了多少本書。

## Queue Introduction

Queue是具有「First-In-First-Out」的資料結構(可以想像成人在排隊)，「最早進入Queue」的人會「最先進餐廳」，「最晚進入Queue」的人則「最晚進餐廳」。


+ Queue的應用是:
    + 被多個程式共享的資源，要排定順序看誰優先執行，可以用Queue。
    + BFS可以用Queue改寫。

## Queue的功能：

- Push(data)：把資料放進Queue。
    + 人排到隊伍中。
- Pop：把「最後面」的資料從Queue中移除。
    + 最後面進隊伍的人，排得不耐煩先離開了。
- getFront：回傳「最前面」的資料，不影響資料結構本身。
    + 確認目前隊伍「最前面」的人是誰。
- getBack：回傳「最後面」的資料，不影響資料結構本身。
    + 確認目前隊伍「最後面」的人是誰。    
- IsEmpty：確認Queue裡是否有資料，不影響資料結構本身。
    + 確認是否還有人在排隊。
- getSize：回傳Queue裡的資料個數，不影響資料結構本身。
    + 記錄目前隊伍中有幾個人。

## 程式碼

這次用Array的概念實作Stack。

```python
class Stack:

    def __init__(self):
        self.stack = []
        
    def push(self, x: int) -> None:
        self.stack.append(x)
    
    def pop(self) -> None:
        self.stack.pop()
        
    def top(self) -> int:
        return self.stack[-1]
    
    def IsEmpty(self):
        if len(self.stack) == 0:
            return True
        else:
            return False
        
    def getSize(self):
        return len(self.stack)
        
```

```python
s = Stack()
s.push(5)
s.push(4)
s.pop()
print(s.top())
print(s.IsEmpty())
print(s.getSize())
```

```python
class Queue:

    def __init__(self):
        self.stack = []

    def push(self, x: int) -> None:
        self.stack.append(x)

    def pop(self) -> int:
        return self.stack.pop(0)
        
    def getFront(self) -> int:
        return self.stack[0]
    
    def getBack(self):
        return self.stack[-1]
        
    def IsEmpty(self) -> bool:
        if len(self.stack) == 0:
            return True
        else:
            return False
    
    def getSize(self):
        return len(self.stack)
```

```python
q = Queue()
q.push(5)
q.push(6)
q.push(4)
q.pop()
print(q.getFront())
print(q.getBack())
print(q.IsEmpty())
print(q.getSize())
```
